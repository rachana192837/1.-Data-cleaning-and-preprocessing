## handling missing data

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data.info()
```

```python
data.drop('Cabin', axis=1, inplace=True)
data.info()
```

## convert into numerical using encoding

```python
data.drop(['Sex','Embarked','Name','Ticket',],axis=1,inplace=True)
```

```python
sex = pd.get_dummies(data['Sex'],drop_first=True).astype(int)
embark= pd.get_dummies(data['Embarked'],drop_first=True).astype(int)
data=pd.concat([data,sex,embark],axis=1)
```

```python
data
```

## standardize numerical features 

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
data
```

```python
data[num_cols].describe()
```

## visualizing using boxplot 

```python

n_cols = 3 
n_rows = (len(num_cols) + n_cols - 1) // n_cols  

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*5, n_rows*4))
axes = axes.flatten()  

for i, col in enumerate(num_cols):
    sns.boxplot(x=data[col], ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')

for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python

def remove_outliers(df, cols):
    df_clean = df.copy()
    for col in cols:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

num_cols = data.select_dtypes(include='number').columns
data_clean = remove_outliers(data, num_cols)

print(f"Original shape: {data.shape}")
print(f"Shape after removing outliers: {data_clean.shape}")
```
